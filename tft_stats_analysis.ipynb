```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = None
```

```python
df = pd.read_csv('tft-data-top100EUW-last20games.csv')
```

Since i got the data from lolchess.gg, i only got informations of the date in the type of "2 days ago", "1 week ago" and so on. So the date is only an approximation.

```python
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
df.describe()
```

```python
date_count = pd.DataFrame(df['date'].value_counts()).sort_index(ascending = False).reset_index().rename(columns={'index': 'Date', 'date': 'Count'})
date_count
```

I will concentrate the analysis on the last 7 days.

```python
df = df[df['date'] >= '2020-12-03']
```

I would like to find typical combinations of traits that often win. Using the whole DataFrame could distort the analysis, because the combinations of the last ranks are not fully built or could have changed over the game. Therefore I try to focues on the ranks 1 to 4. Also I only keep the data with the first and second highest amount of traits to really find "main-combs". 

```python
tdf = df[df['Rank'] <= 4]
tdf.head()
```

```python
ptdf = tdf.iloc[:, 4:]
```

```python
# Start with two highest number(s)

ptdf = ptdf.apply(lambda row: row.nlargest(2),axis=1) 
ptdf.head()
```

```python
combinations = ptdf.notna().dot(ptdf.columns+',').str.rstrip(',')
```

```python
df_test1 = pd.concat([tdf.iloc[:, :4], pd.DataFrame(combinations)], axis=1).rename(columns = {0: 'comb'}) 
df_test1.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
count_combs = pd.DataFrame(df_test1['comb'].value_counts()).reset_index().rename(columns={'index': 'comb', 'comb': 'Count'})
count_combs
```

```python
df_test1 = df_test1[['comb', 'Rank']].groupby('comb').mean().reset_index()
df_test1
```

```python
df_test1 = pd.merge(df_test1, count_combs, on=['comb'])
```

```python
df_test1.sort_values(by = 'Rank').reset_index(drop = True)
```

```python
# Mean Rank of the top 15 often played combs (combs := the two second highest traits, only TOP4)

sorted_df_test1 = df_test1[df_test1['Count'] >= df_test1.nlargest(15, 'Count').iloc[-1]['Count']].sort_values(by = 'Rank')
sorted_df_test1 = sorted_df_test1.rename(columns = {'Rank': 'Mean_Rank', 'comb': 'Main_Combs'})
sorted_df_test1
```

```python
sns.set(font_scale = 1.4)
plt.figure(figsize=(6, 8))

sns.barplot(y = 'Main_Combs', x = 'Mean_Rank', data = sorted_df_test1)
#plt.xticks(rotation=90)
plt.title('Mean Rank of the top 15 often played combs in TOP 4\n\n(combs := highest two traits,\ndata: Last 20 Games of 100 TOP Players in EUW)\n');
```
